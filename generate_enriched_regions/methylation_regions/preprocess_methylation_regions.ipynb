```python
import pandas as pd 
import pathlib 
import os 
import warnings
warnings.filterwarnings("ignore")

savedir = "../../assets/enrichedFeature_pipeline/methylation_regions/processed_methylation_regions"
os.system(f"mkdir -p {savedir}")

bed = {
    "Loyfer_et_al": "/Users/hieunguyen/src/ecd_wgs_enriched_and_kmer_features/resources/Loyfer_atlas/Atlas.U250.l4.hg19.full.tsv",
    "TMD450": "/Users/hieunguyen/src/ecd_wgs_enriched_and_kmer_features/resources/panel.hg19.bed",
    "TSMA": "/Users/hieunguyen/src/ecd_wgs_enriched_and_kmer_features/resources/TSMA.xlsx"
}

##### preprocess Loyfer et al atlas 
df = pd.read_csv(bed["Loyfer_et_al"], sep="\t")
for target in df.target.unique():
    tmpdf = df[df["target"] == target].copy()
    
    tmpdf[["chr", "start", "end"]].to_csv(os.path.join(savedir, f"Loyfer_atlas_{target}.bed"), 
                                        sep="\t", 
                                        index=False, 
                                        header=False)

##### preprocess TSMA 
df = pd.read_excel(bed["TSMA"])
for group in df.Group.unique():
    tmpdf = df[df["Group"] == group].copy()
    tmpdf["chromosome"] = tmpdf["chromosome"].apply(lambda x: f"chr{x}")
    tmpdf[["chromosome", "coord.Start", "coord.End"]].to_csv(os.path.join(savedir, f"TSMA_{group}.bed"), 
                                        sep="\t", 
                                        index=False, 
                                        header=False)
```
