```python
import pandas as pd 
import pathlib
import os 

resourcesdir = "../../resources"
outputdir = "../../assets/enrichedFeature_pipeline/atacseq_bed/hg38"
os.system(f"mkdir -p {outputdir}")

files = [item for item in pathlib.Path(resourcesdir).glob("*.xls")]

for file in files:
    alldf = pd.read_excel(file, sheet_name=None, skiprows=2)
    for sheetname in alldf.keys():
        tmpdf = alldf[sheetname]
        tmpdf = tmpdf.sort_values(by=["seqnames", "start"])
        # print(tmpdf.shape)
        tmpdf[["seqnames", "start", "end"]].to_csv(os.path.join(outputdir, 
                                                                f"{file.name.replace('.xls', f'.{sheetname}.bed')}"), 
                                                   sep="\t", index=False, header=False)
```
